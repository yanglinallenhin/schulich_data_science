```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/yanglinallenhin/data/refs/heads/main/bank_cleaned.csv")

# Drop index column if unnecessary
if "Unnamed: 0" in df.columns:
    df = df.drop(columns=["Unnamed: 0"])

# ---- Basic info ----
print("Dataset shape:", df.shape)
print("\nData types:\n", df.dtypes)
print("\nSummary statistics (numerical):\n", df.describe())

# ---- Check missing values ----
print("\nMissing values per column:\n", df.isnull().sum())

# ---- Target variable distribution ----
target_counts = df["response"].value_counts(normalize=False)
target_ratio = df["response"].value_counts(normalize=True)

print("\nTarget variable distribution (counts):\n", target_counts)
print("\nTarget variable distribution (percentage):\n", target_ratio)

# Plot target distribution
plt.figure(figsize=(6,4))
sns.countplot(x="response", data=df, palette="Set2")
plt.title("Target Variable Distribution (Subscribed Term Deposit)")
plt.xlabel("Response")
plt.ylabel("Count")
plt.show()

# ---- Numerical features distribution ----
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[num_cols].hist(figsize=(12, 10), bins=30, edgecolor="black")
plt.suptitle("Numerical Feature Distributions", fontsize=16)
plt.show()

# ---- Categorical features distribution ----
cat_cols = df.select_dtypes(include=["object"]).drop(columns=["response"]).columns

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=df, palette="Set3", order=df[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
cols=df.select_dtypes(include=['float64','int64'])
cmap1 = 'Blues'
def desc_stats(dataframe):
    desc = dataframe.describe().T
    f,ax = plt.subplots(figsize=(10,
                                 desc.shape[0] * 0.75))
    sns.heatmap(desc,
                annot = True,
                cmap = cmap1,
                fmt= '.2f',
                ax = ax,
                linecolor = 'white',
                linewidths = 1.3,
                cbar = False,
                annot_kws = {"size": 12})
    plt.xticks(size = 14)
    plt.yticks(size = 12,
               rotation = 0)
    plt.title("Descriptive Statistics", size = 14)
    plt.show()
    
desc_stats(cols)
```
